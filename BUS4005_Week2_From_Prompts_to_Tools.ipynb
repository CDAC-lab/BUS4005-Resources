### Paste AI-generated code below
Replace the placeholder in the next code cell with code generated by your AI assistant.


```python

# Paste AI-generated code here and run it.
# Example starter version:

low_reviews = df[df["rating"] <= 3].copy()

theme_keywords = {
    "battery": ["battery"],
    "delivery": ["delivery", "late", "arrived late"],
    "damage": ["damaged", "cracked"],
    "support": ["support", "reply"],
    "setup": ["setup", "instructions", "syncing", "disconnecting"],
    "quality": ["cheap", "leaks", "noisy"]
}

theme_counts = Counter()

for text in low_reviews["review_text"].astype(str):
    lowered = text.lower()
    for theme, keywords in theme_keywords.items():
        if any(k in lowered for k in keywords):
            theme_counts[theme] += 1

print("Complaint themes in low-rated reviews:")
for theme, count in theme_counts.most_common():
    print(f"{theme}: {count}")
```
